```python
class gdregressor:
  def __init__(self,learning_rate=0.01,epochs=100):
    self.coef_=None
    self.intercept_=None
    self.lr=learning_rate
    self.epochs=epochs

  def fit(self,X_train,y_train):
    #initialize the coefficients
    #We initialize the β₀= zero and all the othe coefficient values from β₁ to βₙ are initialized with value =1
    self.intercept_=0
    self.coef_=np.ones(X_train.shape[1])

    for i in self.epochs:
      # finding the predicted value y_hat using the self.intercept_ and self.coef_ values calculated below
      y_hat=np.dot(X_train,self.coef_)+self.intercept_
      # Updating the the β₀ value 
      intercept_der=-2*np.mean(y_train-y_hat)
      self.intercept_=self.intercept_-self.lr*intercept_der
      # Updating the β₁ to βₙ  values 
      coef_der=-2*np.dot((y_train-y_hat),X_train)/X_train.shape[0]
      self.coef_=self.coef_-(self.lr*coef_der)

      # printing final result β₀ value (intercept term) and β₁ to βₙ values (coefficient values for all the columns)
      print(self.intercept_,self.coef_)

  def predict(X_test):
    return np.dot(X_test,self.coef_)+self.intercept_

```
